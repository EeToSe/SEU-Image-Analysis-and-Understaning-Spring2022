# Corner Detection

```python
import cv2
from skimage import io
from matplotlib import pyplot as plt
import numpy as np
```

```python
frame = io.imread('images/corner.jpg',as_gray=True)
frame_blur = cv2.GaussianBlur(frame,(5,5),1)
I_x = cv2.Sobel(frame_blur,-1,1,0,ksize=3)
plt.imshow(np.abs(I_x)/np.abs(I_x.max()),'gray')
```
